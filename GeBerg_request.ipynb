```python
'''
1.標題混入英文字，待處理 / 有一本標題沒有中文，加入規則跳過
2.https://www.gutenberg.org/cache/epub/67976/pg67976-images.html 這本書抓取失敗
3.https://www.gutenberg.org/cache/epub/20968/pg20968-images.html 列表中第70本掛了，要加入規則跳過
'''
```

```python
import requests
import re
import csv
from bs4 import BeautifulSoup

#爬取標題與內文
def scrape_gutenberg_content(id_number):
    url = f"https://www.gutenberg.org/cache/epub/{id_number}/pg{id_number}-images.html"    
    response = requests.get(url)    
    soup = BeautifulSoup(response.text, 'html.parser')

    #提取標題
    title_element = soup.select_one('span#pg-title-no-subtitle')
    title = title_element.get_text(strip=False)

    #提取內文
    paragraphs = []
    i = 1
    while True:
        #內文段落ID
        paragraph_id = f'p#id{str(i).zfill(5)}'
        paragraph_element = soup.select_one(paragraph_id)
        
        #如果找不到段落就break
        if not paragraph_element:
            break
        
        #移除英文
        text = paragraph_element.get_text(strip=False)
        text_without_english = re.sub(r'[a-zA-Z]', '', text)
        
        paragraphs.append(text_without_english)
        i += 1

    #每個段落空一行
    content = '\n\n'.join(paragraphs)

    #存成txt檔
    with open(f'./project_gutenberg/{title}.txt', 'w', encoding='utf-8') as f:
        f.write(content)

    print(f"文件已成功保存")

#讀取中文書的ID列表
def process_csv(csv_path):
    with open(csv_path, 'r', encoding='utf-8') as csvfile:        
        reader = csv.reader(csvfile)        
        for row in reader:            
            id_number = row[0]
            scrape_gutenberg_content(id_number)


#執行
csv_path = './book_ids.csv' 
process_csv(csv_path)

```
